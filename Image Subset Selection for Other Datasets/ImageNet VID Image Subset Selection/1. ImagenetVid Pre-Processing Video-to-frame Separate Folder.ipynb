```python
import cv2
import os

# Define the directory path where your 'train' folder is located
input_dir_path_train = './vidvrd-dataset/videos_with_one_category_2/train'
output_dir_path_train = './vidvrd-dataset/videos_with_one_category_img_6FPS_sep/train'

input_dir_path_val = './vidvrd-dataset/videos_with_one_category_2/val'
output_dir_path_val = './vidvrd-dataset/videos_with_one_category_img_6FPS_sep/val'
```

```python
def convert_mp4_to_jpeg(input_dir_path, output_dir_path):
    # Loop over all subdirectories in the 'train' directory
    for subdir in os.listdir(input_dir_path):
        subdir_path = os.path.join(input_dir_path, subdir)
        
        # Skip if not a directory
        if not os.path.isdir(subdir_path):
            continue
        
        # Create a corresponding subdirectory in the output directory
        output_subdir_path = os.path.join(output_dir_path, subdir)
        os.makedirs(output_subdir_path, exist_ok=True)
        
        # Loop over all MP4 files in each subdirectory
        for file in os.listdir(subdir_path):
            if file.endswith('.mp4'):
                file_path = os.path.join(subdir_path, file)
                video_name = os.path.splitext(file)[0]
                
                # Open the video file
                vidcap = cv2.VideoCapture(file_path)
                success, image = vidcap.read()
                count = 0
                
                while success:
                    if count % 5 == 0:
                        output_folder_path = os.path.join(output_subdir_path, video_name)
                        os.makedirs(output_folder_path, exist_ok=True)
                        frame_file_path = os.path.join(output_folder_path, f"{video_name}_{count:04d}.jpg")
                        cv2.imwrite(frame_file_path, image)
                        print(f'Saved frame {count} from video {file_path}')
                    
                    success, image = vidcap.read()
                    count += 1
```

```python
convert_mp4_to_jpeg(input_dir_path_train, output_dir_path_train)
```
